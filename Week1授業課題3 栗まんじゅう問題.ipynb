```python
SPECE = (27/2)*10**78
TOKYO_DOME = 12400000

F = TOKYO_DOME

def baibai(F):
    KURI_MANZYUU = 0.0001
    KURI_MANZYUU_baibai = 0.0001
    
    count = 0
    KURI_MANZYUU_baibai_list = []
    while KURI_MANZYUU_baibai < F:
        KURI_MANZYUU_baibai = KURI_MANZYUU_baibai*2
        count +=5
        KURI_MANZYUU_baibai_list.append(KURI_MANZYUU_baibai)
    return KURI_MANZYUU_baibai_list
    return count
print(baibai(SPECE))
print('{:.1f}時間かかります'.format(count/60))


import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(baibai(SPECE), color = "orange", marker='o', label='SPECE')
plt.plot(baibai(TOKYO_DOME), color = "blue", marker='o', label='TOKYO_DOME')
plt.legend()
plt.show()
```
