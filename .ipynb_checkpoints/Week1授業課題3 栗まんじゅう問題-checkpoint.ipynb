Week 1: 栗まんじゅう問題
----

SPECEの定義
１光年を約10兆km＝1013km＝1016ｍとして考える。

すると150億光年は、1.5×1010光年＝1.5×1010×1016＝(3/2)×1026ｍ

ゆえに、宇宙の体積は、(4/3)π×{(3/2)×1026}3≒　(27/2)×10 78(乗) ｍ3


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def baibai(F):
    KURI_MANZYUU_baibai = 0.0001
    count_initial = 0
    count = []
    KURI_MANZYUU_baibai_list = []
    while KURI_MANZYUU_baibai < F:
        KURI_MANZYUU_baibai = KURI_MANZYUU_baibai*2
        KURI_MANZYUU_baibai_list.append(KURI_MANZYUU_baibai)
        count_initial+=5
        count.append(count_initial)
    return KURI_MANZYUU_baibai_list

SPECE = (27/2)*10**78
TOKYO_DOME = 12400000

print(baibai(SPECE))




```

太陽系が栗まんじゅうで覆われるのは何分後でしょうか
---

栗まんじゅうが宇宙を覆ってしまう時のグラフ
----

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(baibai(SPECE), color = "orange", marker='o', label='SPECE')
plt.legend()
plt.show()
```

```python
print('{:.1f}分間かかります'.format(len(baibai(SPECE))))
```

東京ドームが一杯になる時間
---

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(baibai(TOKYO_DOME)[:-1], color = "blue", marker='o', label='TOKYO_DOME')
plt.legend()
plt.show()
```

```python
print('{:.1f}分間かかります'.format(len(baibai(TOKYO_DOME))))
```
